```python
import nltk
nltk.download('vader_lexicon')
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer

#Creamos una instancia para el analizador de sentimientos de Vader
sentiment_analyzer = SentimentIntensityAnalyzer()

def categoriza_sentimiento(texto):
    #Función que clasifica los sentimientos en base al calculo
    # del tono emocional de los comentarios y les asigna un valor de
    # 0 (malo), 1 (neutral) o 2 (positivo)
    retorno = 1 # Inicia una variable con el valor para la categoría de Neutro
    if(len(texto) > 0):
        sentimiento = sentiment_analyzer.polarity_scores(texto)["compound"]
        if(sentimiento >= 0.33):
           retorno = 2 #Categoría de Positivo
        elif(sentimiento <= -0.33):
            retorno = 0 #Categoría de Negativo

    return retorno

Df_desanidado_reviews["sentiment_analysis"] = Df_desanidado_reviews["review"].apply(categoriza_sentimiento)
```

```python
def PlayTimeGenre(genres: str):
    """
    Devuelve el año con más horas jugadas para un género determinado.

    Args:
        genero: El género de los juegos a analizar.

    Returns:
        Un diccionario con el año con más horas jugadas y el número de horas jugadas.
    """

    # Importar las librerías necesarias
    import pandas as pd
    import datetime

    # Leer el archivo CSV con los datos de los juegos
    df = pd.read_csv("/content/dataISO.csv")

    # Filtrar los juegos por género
    df = df[df["genres"] == genres]

    # Calcular las horas jugadas por año
    played_hours_per_year = df.groupby(df["release_date"])["playtime_forever"].sum()

    # Obtener el año con más horas jugadas
    max_hours_year = played_hours_per_year.sort_values(ascending=False).index[0]

    # Devolver el resultado
    return {"Año de lanzamiento con más horas jugadas para Género {}".format(genres): max_hours_year}
```
